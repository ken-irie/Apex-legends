```python
import cv2
import matplotlib.pyplot as plt
import os.path
import math
import openpyxl
```

```python
def get_distance(x1, y1, x2, y2):
 if __name__ == '__main__':  
    d = math.sqrt((x2 - x1) ** 2 + (y2 - y1) ** 2)
    return d
```

```python
path = 'C:\\Users\\kenta\\Desktop\\apex\\jissou\\DataseteName'
name = 'sample_video_img_'
types = '.jpg'
cascade = cv2.CascadeClassifier("C:\\Users\\kenta\\Desktop\\apex\\jissou\\opencvcascade2\\cascade\\cascade.xml")
path2 = 'C:\\Users\\kenta\\Desktop\\apex\\jissou\\find_mato_center'
img_x_center = 640
img_y_center = 360

book = openpyxl.Workbook()

sum = 0

for i in range(100):
    i = str(i)
    i = i.zfill(3)
    name_iamge = name + i + types
    image_path = os.path.join(path, name_iamge)
    img = cv2.imread(image_path)
#     cv2.imshow('image_000', img)
#     cv2.waitKey(0)
#     cv2.destroyAllWindows()
    print(name_iamge)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    mato = cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=3, minSize=(30, 30))
    new_name = 'result_' + i +'.jpg'
    mato_path = os.path.join(path2, new_name)
#    print(mato_path)
    #顔領域を赤色の矩形で囲む
    for (x, y, w, h) in mato:
        if 500 < x < 700 and 300 < y < 390:
            cv2.rectangle(img, (x, y), (x + w, y+h), (0,0,200), 3)
#            print(x,y)
            x_center = int(x) + int(w)/2
            y_center = int(y) + int(h)/2
#            print(int(x) + int(w)/2, int(y) + int(h)/2)
            if __name__ == '__main__':
                d = get_distance(x_center, y_center, img_x_center, img_y_center)
                print(d)
                d = round(d)
                



                sum = sum + d
                print(sum)


#     # 結果画像を保存
#           cv2.imwrite(mato_path, img)


# #結果画像を表示
# #         cv2.imshow('image_000', img)
# #         cv2.waitKey(0)
# #         cv2.destroyAllWindows()
```
